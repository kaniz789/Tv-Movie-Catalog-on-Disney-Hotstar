# **Calculates the frequency of values in the 'genre' column**

```python
data = Movie['genre'].value_counts().sort_values(ascending=False)[:10]
colors = sns.color_palette('husl', len(data))
sns.barplot(x=data.index, y=data, palette=colors)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show;
```

```python
data = tv_shows['genre'].value_counts().sort_values(ascending=False)[:10]
colors = sns.color_palette('husl', len(data))
sns.barplot(x=data.index, y=data, palette=colors)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show;
```

# **distribution of content across different age ratings**

```python
data = Movie['age_rating'].value_counts()
colors = sns.color_palette('husl', len(data))
sns.barplot(x=data.index, y=data, palette=colors)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Movie and TV Show Genres over the Years**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Movie, x='year', y='genre', label='Movie', alpha=0.6)
sns.scatterplot(data=tv_shows, x='year', y='genre', label='TV Show', alpha=0.6)
plt.title('Scatter plot of Movie and TV Show Genres over the Years')
plt.xlabel('Year')
plt.ylabel('Genre')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
