# **4.** Calculates the frequency of values in the 'genre' column

*  Calculating the frequency of values in the 'genre' column helps identify the most common types of content Disney+Hotstar produces.

*  This reveals Disney+Hotstar's dominant genre preferences and can highlight trends or shifts in content focus over time.

```python
data = Movie['genre'].value_counts().sort_values(ascending=False)[:10]
colors = sns.color_palette('husl', len(data))
plt.title('Top 10 Movie Genres on Disney+ Hotstar')
sns.barplot(x=data.index, y=data, palette=colors)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show;
```

```python
data = tv_shows['genre'].value_counts().sort_values(ascending=False)[:10]
colors = sns.color_palette('husl', len(data))
plt.title('Top 10 TV_shows Genres on Disney+ Hotstar')
sns.barplot(x=data.index, y=data, palette=colors)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show;
```

# **6.** distribution of age_rating

*  Examining the distribution of age ratings provides insight into the target audiences Disney focuses on.
* It helps us understand whether the majority of content is family-friendly, or if there's a growing presence of content aimed at teens and adults.

```python
data = Movie['age_rating'].value_counts()
colors = sns.color_palette('husl', len(data))
plt.title('Distribution of Age Ratings for Movies')
sns.barplot(x=data.index, y=data, palette=colors)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data = tv_shows['age_rating'].value_counts()
colors = sns.color_palette('husl', len(data))
plt.title('Distribution of Age Ratings for TV_shows')
sns.barplot(x=data.index, y=data, palette=colors)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **7.** Movie and TV Show Genres over the Years

*  Analyzing movie and TV show genres over the years reveals how Disney+Hotstar's content themes have evolved with changing audience preferences.
*  This helps identify trends, such as the rise or decline of specific genres, and whether certain genres are more common in movies or TV shows over time.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Movie, x='year', y='genre', label='Movie', alpha=0.6)
sns.scatterplot(data=tv_shows, x='year', y='genre', label='TV Show', alpha=0.6)
plt.title('Scatter plot of Movie and TV Show Genres over the Years')
plt.xlabel('Year')
plt.ylabel('Genre')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
